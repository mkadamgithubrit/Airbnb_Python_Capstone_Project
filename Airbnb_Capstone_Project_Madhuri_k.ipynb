# **Data Analysis on Airbnb Dataset**

##Q.1 What can we learn about different hosts and areas?

```python
# Distribution of Hosts and Listings
host_count = df['host_id'].nunique()
listing_count = df.shape[0]
# The top 10 popular neighborhoods based on the number of listings.
top_hosts = df.groupby(['host_name'])['calculated_host_listings_count'].max().reset_index()
top_hosts.sort_values(by='calculated_host_listings_count', ascending=False).head(10)
```

#### the top 10 hosts with the highest number of listings, making it easy to identify the most active hosts in the dataset.

##Q.2 Correlation Heatmap

```python
numeric_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Pastel1', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

####The heatmap presents a color-coded representation of the correlation coefficients, with warmer colors indicating positive correlations and cooler colors representing negative correlations.

##Q.3 Display the top property types with the highest number of listings

```python

# Group the data by 'room_type' and calculate the total number of listings for each room type
property_bookings = df.groupby('room_type')['calculated_host_listings_count'].sum().reset_index()

# Sort the data by total number of listings in descending order
property_bookings = property_bookings.sort_values(by='calculated_host_listings_count', ascending=False)

# Display the top  property types with the highest number of listings
top_property_types = property_bookings.head(5)

# Create a bar plot to visualize the top property types with the highest number of listings
plt.figure(figsize=(10, 6))
plt.bar(top_property_types['room_type'], top_property_types['calculated_host_listings_count'], color='pink')
plt.title('Top Property Types with the Highest Number of Listings')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

####we analyzed the data to identify the top property types with the highest number of listings. The results show that [room type 1] and [room type 2] are the most popular choices among hosts.

##Q.4 Which are the different neighborhood groups present in the dataset, and what percentage of the data does each group represent?

```python
# Group the data by 'neighbourhood_group' and count the occurrences
group_counts = df['neighbourhood_group'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(group_counts, labels=group_counts.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Distribution of Data by Neighbourhood Group')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

####The pie chart illustrates the distribution of data across different neighborhood groups. It provides an overview of the relative prevalence of each group, helping identify the dominant neighborhoods and their representation in the dataset.

##Q.5 How does the price distribution of Airbnb listings vary based on different room types and across various neighborhood groups (e.g., boroughs, districts)?

```python
# creat the bar chart
sns.barplot(data=df, x="room_type", y="price", hue="neighbourhood_group")
plt.title('Price Distribution by Room Type and Neighbourhood Group')
plt.xlabel('Room Type')
plt.ylabel('Price')

plt.show()
```

####In this visualization, we can observe how prices vary across room types and how different neighborhood groups influence the price distribution for each room type.






##Q.6 Creates a box plot for variables to visualize their distributions and identify outliers

```python

# Selecting the numerical columns for the box plot
numerical_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

# Creating the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_columns])
plt.title('Box Plot of Numerical Variables')
plt.ylabel('Values')
plt.xticks(rotation=15, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

####The output will be a single box plot with boxes representing the interquartile range (IQR), whiskers extending to data points within a certain range, and any individual points beyond the whiskers considered outliers for each of the selected numerical columns. The x-axis will display the column names, and the y-axis will represent the values of these numerical variables.

##Q.7 How do the prices differ among different room types (e.g., entire homes, private rooms, shared rooms)?

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='room_type', y='price', palette='Set2')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```

#### a bar plot with the x-axis representing the different room types ('room_type'), the y-axis representing the average price ('price'), and bars of different heights for each room type. basical The code creates a bar plot to display the average price for each room type in the dataset

##Q.8 Which hosts are the busiest and why?


```python
busy_hosts = df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busy_hosts.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
name = busiest_hosts['host_name']
reviews = busiest_hosts['number_of_reviews']

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(name, reviews, color ='skyblue',
        width = 0.4)

plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts")
plt.show()
```

####The code creates a bar plot using Matplotlib, where the x-axis represents the "Name of the Host" and the y-axis represents the "Number of Reviews." It visualizes the busiest hosts and their corresponding review counts, helping to identify which hosts have received the most reviews.

##Q.9 Top 5 areas with most number of bookings.

```python
# Group by 'neighbourhood_group' and calculate the number of bookings for each group
top_neighbourhood_groups = df['neighbourhood_group'].value_counts().nlargest(5).index.tolist()

# Filter the DataFrame to include data for the top 5 neighbourhood groups
top_neighbourhood_groups_data = df[df['neighbourhood_group'].isin(top_neighbourhood_groups)]

# Create the scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=top_neighbourhood_groups_data, x='longitude', y='latitude', hue='neighbourhood_group', palette='Set1', s=80)
plt.title('Top 5 Neighbourhood Groups with Most Number of Bookings (Scatter Plot)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###Groups the data by 'neighbourhood_group' and calculates the number of bookings for each group, selecting the top 5 groups with the most bookings.Filters the DataFrame to include data only for the top 5 neighbourhood groups.

##Q.10 top 10 areas with most numbers of visitors.

```python
# Group the data by 'neighbourhood_group' and 'neighbourhood' and count the occurrences
visitors_count_by_area = df.groupby(['neighbourhood_group', 'neighbourhood']).size().reset_index(name='count')

# Sort the data by the number of visitors in descending order
top_10_areas = visitors_count_by_area.sort_values(by='count', ascending=False).head(10)

# Bar chart to visualize the top 10 areas with the most number of visitors by neighbourhood_group
plt.figure(figsize=(8, 6))
plt.bar(top_10_areas['neighbourhood_group'] + ': ' + top_10_areas['neighbourhood'], top_10_areas['count'])
plt.xlabel('Neighborhood Group: Neighborhood')
plt.ylabel('Number of Visitors')
plt.title('Top 10 Areas with the Most Number of Visitors by Neighbourhood Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### the top 10 areas with the most visitors, grouped by neighborhood groups. Each bar represents a neighborhood within a neighborhood group, and its height corresponds to the "Number of Visitors." The x-axis labels are formatted as "Neighborhood Group: Neighborhood," and the chart is titled "Top 10 Areas with the Most Number of Visitors by Neighbourhood Group.
